```python
import logging
logging.basicConfig(filename = 'ravi.log' , level = logging.INFO)

import sys
logging.info("exception hierarchy function is created")
def print_exception_hierarchy():
    logging.info("exception subclasses list is created")
    exception_classes = []
    for name, obj in sys.modules['builtins'].__dict__.items():
        if isinstance(obj, type) and issubclass(obj, BaseException):
            
            exception_classes.append(obj)
            logging.info("objects are appended in exception classes ")
    print("Python Exception Hierarchy:")
    print("---------------------------")
    print(BaseException.__name__)
    print("|")
    for exception_class in exception_classes:
        print("|--", exception_class.__name__)
        print("    |")
        for subclass in exception_class.__subclasses__():
            print("    |--", subclass.__name__)
    print()

logging.info(" comand is given to obtain hierarchy of exceptions")
print_exception_hierarchy()
```

'ArithmaticError' class is the base class related to aithmatic in python.
it serves as the parent class for all arithmatic exceptions.

Here are basically two common types of 'Arithmatic' exception
1.- ZeroDivisionError
2.- OverFlowError

```python

import logging
logging.basicConfig(filename = "ravi.log" , level = logging.INFO)
logging.info("a and b ; two variables are created")
a =10
b = 0
logging.info("try statement is applied to handle exception")
try :
    a/b
    
except  ZeroDivisionError as e:
    logging.error("exception is handled".format(e))
```
